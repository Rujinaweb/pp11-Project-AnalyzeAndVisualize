```python
#Display students who have a passing year before 2020
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("student_data.csv")

df['passing_year'] = pd.to_numeric(df['passing_year'], errors='coerce')
passing_students_before_2020 = df[(df['passing_year'] < 2020) & (df['passing_year'].notna())]

passing_students_before_2020
```

```python
#Calculate the median attendance for students in the "Architecture" department.
import pandas as pd
architecture_students = df[df['department'] == 'Architecture']
median_attendance = architecture_students['attendance'].median()

print(f"The median attendance for students in the Architecture department is: {median_attendance}")
```

```python
#Calculate the median attendance for students in the "Business" department.
import pandas as pd
business_students = df[df['department'] == 'Business']
median_attendance = business_students['attendance'].median()

print(f"The median attendance for students in the Business department is: {median_attendance}")
```

```python
#Replace missing values in the "passing_year" column with the most frequent passing year.
import pandas as pd

df_fillna=df.fillna({"passing_year":"none"})
df.isnull().sum()



```

```python
#Create a bar chart comparing attendance percentages between different departments.
import matplotlib.pyplot as plt
attendance_percentages = df.groupby('department')['attendance'].mean()
plt.figure(figsize=(10, 6))
plt.bar(attendance_percentages.index, attendance_percentages.values)
plt.xlabel("Department")
plt.ylabel("Average Attendance Percentage")
plt.title("Comparison of Attendance Percentages Across Departments")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
